```python
import pandas as pd
from PIL import Image
import io
```

```python
def get_file_path(filename, label):
    result = "body_posture_dataset/"+str(label)+"/"+filename
    return result
```

```python
train1= pd.read_csv("body_posture_dataset/train_1.csv")
train2= pd.read_csv("body_posture_dataset/train_2.csv")
train3= pd.read_csv("body_posture_dataset/train_3.csv")
test1= pd.read_csv("body_posture_dataset/test_1.csv")
test2= pd.read_csv("body_posture_dataset/test_2.csv")
test3= pd.read_csv("body_posture_dataset/test_3.csv")
```

```python
LABEL = 3
SAVE_PATH = "cropped_dataset/train/3/"
FILE = train3
```

```python
filenames = [ row["filename"] for index, row in FILE.iterrows()]
```

```python
bboxs =[ 
    [row["xmin"], row["ymin"], row["xmax"], row["ymax"]] 
    for index, row in FILE.iterrows()  
]
```

```python
for index, filename in enumerate(filenames):
    img = Image.open(get_file_path(filename, LABEL))
    crop_img = img.crop(bboxs[index])
    crop_img.save(SAVE_PATH+filename)
```
