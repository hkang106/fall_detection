```python
from PIL import Image
from glob import glob
import os
```

```python
file_paths = glob("dataset_8labels/crawling/*.png")
```

```python
file_paths
```

```python
for path in file_paths:
    img = Image.open(path)
    img.save("dataset_8labels/crawling/jpg/"+os.path.basename(path).replace(".png", ".jpg"))
```
